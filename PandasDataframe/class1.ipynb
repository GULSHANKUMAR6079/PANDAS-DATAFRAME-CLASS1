```python
def did_kohli_played(players_list):
    return "V Kohli" in players_list
```

```python
did_kohli_played(["V Kohli","rohit sharma","ajay devgan"])
```

```python
did_kohli_played(["sunny deol","rohit sharma","ajay devgan"])
```

```python
ipl["all_players"].apply(did_kohli_played)
```

```python
ipl["did_kohli_played"]=ipl["all_players"].apply(did_kohli_played)
ipl["did_kohli_played"]
```

```python
ipl[ipl["did_kohli_played"]]
# it will give me al the  Tryue values 
```

```python
ipl[(ipl["City"]== "Delhi")&(ipl["did_kohli_played"]==True)]
# it will give me the value of that rows only where city ==Delho and Kohli played that match
```

```python
#as the dataframe is already sorted and mereko chahiye ki last or recent match kon sa tha as we know ki sara match Delhi mein hua tha and kohli sab mein khela tha toh hum log drop_duplicates use kar sakte hai and keep ="first" kar denge taki recent wala match ka hi o/p de 
ipl[(ipl["City"]== "Delhi")&(ipl["did_kohli_played"]==True)].drop_duplicates(subset=["City","did_kohli_played"], keep="first")
```
